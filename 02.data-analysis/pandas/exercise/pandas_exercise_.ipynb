## 문제 5. 결측치 처리

`df`에 결측치를 처리하기 위해 `df_clean`으로 데이터를 복했습니다. `df_clean`에 대해서 결측치가 있는 컬럼을 확인한 뒤, 다음 방식으로 결측치를 처리하세요. 수치형 컬럼은 `num_cols`, 문자형 컬럼은 `cat_cols`를 사용하세요.

- 전체 결측치의 개수를 확인하세요. (754개)
- 수치형 컬럼의 결측치는 각 컬럼의 중앙값으로 채우세요.
- 문자형 컬럼의 결측치는 `"Unknown"`으로 채우세요.
- 결측치 처리 후 `df_clean`에 결측치가 남아 있는지 다시 확인하세요.

```python
df_clean = df.copy()
num_cols = df_clean.select_dtypes(include="number").columns
cat_cols = df_clean.select_dtypes(exclude="number").columns

print("결측치 총합:", df.isnull().sum().sum())

df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())
df_clean[cat_cols] = df_clean[cat_cols].fillna("Unknown")

print("남은 결측치 총합:", df_clean.isna().sum().sum())
```

## 문제 6. IQR 방식으로 이상치 찾기와 제거하기

`df_clean`에 대해서 `Particle_Count` 컬럼을 대상으로 IQR 방식의 이상치 기준을 계산하세요.

다음 값을 구하세요.

- Q1, Q3, IQR, lower bound, upper bound

그 다음 다음 작업을 수행하세요.

- `Particle_Count`가 IQR 기준을 벗어난 행을 선택해서 `outlier_df`에 대입하고 행수를 출력하세요. 61행

- 이상치를 제거한 결과를 `df_no_outlier`에 저장하고 행수를 확인하세요. `no_outlier_df`의 행수는 3539가 되야함

```python
col = "Particle_Count"

# qauntail() 함수를 사용해서 Q1, Q3, IQR를 구하세요.
Q1 = df_clean[col].quantile(0.25)
Q3 = df_clean[col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("lower bound:", lower_bound)
print("upper bound:", upper_bound)

# Particle_Count가 IQR 기준을 벗어난 행만 필터링
outlier_cond = (df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)
outlier_df = df_clean[outlier_cond]
print("이상치 포함 데이터 크기:", outlier_df.shape)


no_outlier_df = df_clean[~outlier_cond].copy()
print("이상치 제거 후 데이터 크기:", no_outlier_df.shape)
```

## 문제 7. 공장별 불량률 계산

`df_clean`에 대해서 공장별(`Factory`)로 다음 값을 계산하세요. hint: groupby해서 size, sum, mean을 사용하세요.

- 생산 건수
- 불량 건수
- 불량률

계산한 정보를 담은 `df_factory` 데이터프레임에 대해서 불량률이 높은 순서로 정렬하고 어떤 공장의 불량률이 가장 높은지 확인하세요. C공장

```python
df_size = df_clean.groupby('Factory').size()
df_defect = df_clean.groupby('Factory')['Defect'].sum()
df_defect_rate = df_clean.groupby('Factory')['Defect'].mean()

display(df_size)
display(df_defect)
display(df_defect_rate)

df_factory = pd.concat([df_size,df_defect,df_defect_rate], axis=1)
df_factory.columns = ['Count', 'Defect Count', 'Defect Rate']
df_factory.sort_values('Defect Rate', ascending=False)
```

## 문제 8. 공장 × 교대조, 공장별 생산건수 계산

'df_clean'에 대해 다음 작업을 수행하세요.

- `pivot_table()`을 사용하여 공장(Factory) × 교대조별(Shift) 불량률을 계산하여 `f_shift_rate`에 저장하고 확인하세요.
- `crosstab()`을 사용하여 공장별 (Product_Type) 생산 건수를 `f_product_count`에 저장하고 확인하세요.

```python
f_shift_rate = pd.pivot_table(
    df_clean,
    index="Factory",
    columns="Shift",
    values="Defect",
    aggfunc="mean"
)
display(f_shift_rate)


f_product_count = pd.crosstab(df_clean["Factory"], df_clean["Product_Type"])
display(f_product_count)
```
